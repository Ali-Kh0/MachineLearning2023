**Problem 1:** Create a NumPy array of 20 random integers between 1 and 100. Calculate the mean, median, and standard deviation of the array.

````python
import numpy as np

random_array = np.random.randint(1, 101, 20)


mean = np.mean(random_array)
median = np.median(random_array)
std_deviation = np.std(random_array)

print("Random Array:", random_array)
print("Mean:", mean)
print("Median:", median)
print("Standard Deviation:", std_deviation)
````

**Problem 4:** Create two NumPy arrays, 'arr1' and 'arr2', both of size 5x5 with random integer values. Perform element-wise addition, subtraction, multiplication, and division between the two arrays.

````python
import numpy as np

# Create two random 5x5 NumPy arrays
arr1 = np.random.randint(1, 100, (5, 5))
arr2 = np.random.randint(1, 100, (5, 5))

# Perform element-wise addition
addition_result = arr1 + arr2

# Perform element-wise subtraction
subtraction_result = arr1 - arr2

# Perform element-wise multiplication
multiplication_result = arr1 * arr2

# Perform element-wise division
# We'll add 1 to the denominator to avoid division by zero
division_result = arr1 / (arr2 + 1)

# Print the original arrays and the results
print("Array 1:")
print(arr1)
print("\nArray 2:")
print(arr2)
print("\nElement-wise Addition:")
print(addition_result)
print("\nElement-wise Subtraction:")
print(subtraction_result)
print("\nElement-wise Multiplication:")
print(multiplication_result)
print("\nElement-wise Division:")
print(division_result)
````

**Problem 10:** Given a Pandas DataFrame containing information about students, add a new column 'Grade' based on the following criteria:
* If 'Score' >= 90, Grade is 'A'
* If 'Score' >= 80, Grade is 'B'
* If 'Score' >= 70, Grade is 'C'
* If 'Score' < 70, Grade is 'D'

````python
import pandas as pd

# Sample DataFrame with student information
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Score': [95, 78, 60, 85, 72]
}

student_df = pd.DataFrame(data)

# Function to calculate grades
def calculate_grade(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    else:
        return 'D'

# Add a new 'Grade' column based on the 'Score' column
student_df['Grade'] = student_df['Score'].apply(calculate_grade)

# Display the updated DataFrame
print(student_df)
````

**Problem 11:** You are given a dataset containing monthly sales data for three products (A, B, and C) over a two-year period. Your task is to perform various data analysis tasks using NumPy, Pandas, and Matplotlib.


*   Dataset:

```
import pandas as pd

data = {
    'Month': pd.date_range(start='2021-01-01', periods=24, freq='M'),
    'Product A Sales': [500, 480, 600, 750, 900, 850, 920, 1100, 1300, 1350, 1500, 1450, 1550, 1600, 1650, 1600, 1500, 1400, 1600, 1700, 1800, 1750, 1850, 1900],
    'Product B Sales': [300, 320, 400, 450, 500, 580, 700, 750, 820, 900, 950, 980, 1050, 1100, 1150, 1200, 1250, 1300, 1350, 1400, 1500, 1600, 1550, 1700],
    'Product C Sales': [200, 210, 250, 280, 320, 350, 380, 400, 420, 440, 460, 480, 500, 520, 540, 560, 580, 600, 620, 640, 660, 680, 700, 720]
}

sales_df = pd.DataFrame(data)
```
*  Tasks:
1. Calculate and display the total sales for each product over the two-year period.

2. Calculate and display the average monthly sales for each product.

3. Find the month with the highest sales for each product, and display the product and the sales value for that month.

4. Calculate and display the percentage change in sales for each product from January to December in the second year (2022).

5. Create a line chart using Matplotlib to visualize the monthly sales data for each product over the two-year period. Label the axes and provide a title for the chart.

6. Calculate the correlation between Product A and Product B sales. Is there a strong correlation between these two products? Display the correlation coefficient.


````python
import pandas as pd
import matplotlib.pyplot as plt

# Create the DataFrame
data = {
    'Month': pd.date_range(start='2021-01-01', periods=24, freq='M'),
    'Product A Sales': [500, 480, 600, 750, 900, 850, 920, 1100, 1300, 1350, 1500, 1450, 1550, 1600, 1650, 1600, 1500, 1400, 1600, 1700, 1800, 1750, 1850, 1900],
    'Product B Sales': [300, 320, 400, 450, 500, 580, 700, 750, 820, 900, 950, 980, 1050, 1100, 1150, 1200, 1250, 1300, 1350, 1400, 1500, 1600, 1550, 1700],
    'Product C Sales': [200, 210, 250, 280, 320, 350, 380, 400, 420, 440, 460, 480, 500, 520, 540, 560, 580, 600, 620, 640, 660, 680, 700, 720]
}

sales_df = pd.DataFrame(data)

# Task 1: Calculate and display the total sales for each product over the two-year period.
total_sales = sales_df.sum()
print("Total Sales for Each Product:")
print(total_sales[['Product A Sales', 'Product B Sales', 'Product C Sales']])
print()

# Task 2: Calculate and display the average monthly sales for each product.
average_sales = sales_df.mean()
print("Average Monthly Sales for Each Product:")
print(average_sales[['Product A Sales', 'Product B Sales', 'Product C Sales']])
print()

# Task 3: Find the month with the highest sales for each product and display the product and the sales value for that month.
max_sales_month_A = sales_df[sales_df['Product A Sales'] == sales_df['Product A Sales'].max()]['Month'].values[0]
max_sales_value_A = sales_df['Product A Sales'].max()
print(f"Month with Highest Sales for Product A: {max_sales_month_A}, Sales Value: {max_sales_value_A}")

max_sales_month_B = sales_df[sales_df['Product B Sales'] == sales_df['Product B Sales'].max()]['Month'].values[0]
max_sales_value_B = sales_df['Product B Sales'].max()
print(f"Month with Highest Sales for Product B: {max_sales_month_B}, Sales Value: {max_sales_value_B}")

max_sales_month_C = sales_df[sales_df['Product C Sales'] == sales_df['Product C Sales'].max()]['Month'].values[0]
max_sales_value_C = sales_df['Product C Sales'].max()
print(f"Month with Highest Sales for Product C: {max_sales_month_C}, Sales Value: {max_sales_value_C}")
print()

# Task 4: Calculate and display the percentage change in sales for each product from January to December in the second year (2022).
january_sales_A = sales_df[sales_df['Month'] == '2022-01-31']['Product A Sales'].values[0]
december_sales_A = sales_df[sales_df['Month'] == '2022-12-31']['Product A Sales'].values[0]
percentage_change_A = ((december_sales_A - january_sales_A) / january_sales_A) * 100

january_sales_B = sales_df[sales_df['Month'] == '2022-01-31']['Product B Sales'].values[0]
december_sales_B = sales_df[sales_df['Month'] == '2022-12-31']['Product B Sales'].values[0]
percentage_change_B = ((december_sales_B - january_sales_B) / january_sales_B) * 100

january_sales_C = sales_df[sales_df['Month'] == '2022-01-31']['Product C Sales'].values[0]
december_sales_C = sales_df[sales_df['Month'] == '2022-12-31']['Product C Sales'].values[0]
percentage_change_C = ((december_sales_C - january_sales_C) / january_sales_C) * 100

print("Percentage Change in Sales from January to December 2022:")
print(f"Product A: {percentage_change_A:.2f}%")
print(f"Product B: {percentage_change_B:.2f}%")
print(f"Product C: {percentage_change_C:.2f}%")
print()

# Task 5: Create a line chart using Matplotlib to visualize the monthly sales data for each product over the two-year period. Label the axes and provide a title for the chart.
plt.figure(figsize=(10, 6))  # Set the figure size

plt.plot(sales_df['Month'], sales_df['Product A Sales'], label='Product A', marker='o')
plt.plot(sales_df['Month'], sales_df['Product B Sales'], label='Product B', marker='o')
plt.plot(sales_df['Month'], sales_df['Product C Sales'], label='Product C', marker='o')

plt.xlabel('Month')  # X-axis label
plt.ylabel('Sales')  # Y-axis label
plt.title('Monthly Sales Over Two Years')  # Title
plt.legend()  # Add legend

plt.grid(True)  # Add grid lines (optional)

plt.show()

# Task 6: Calculate the correlation between Product A and Product B sales. Display the correlation coefficient.
correlation_AB = sales_df['Product A Sales'].corr(sales_df['Product B Sales'])
print(f"Correlation between Product A and Product B Sales: {correlation_AB:.2f}")
````
